```python
import pandas as pd
import numpy as np
import osmnx as ox, networkx as nx, geopandas as gpd, matplotlib.pyplot as plt
from geopandas import GeoDataFrame
from shapely.geometry import Point
from shapely.geometry import Polygon
import shapely.wkt
import datetime
%matplotlib inline
```

```python
stops = pd.read_csv('RTD-GTFS/stops.csv')
```

```python
#Stockton Blocks
BGs = gpd.read_file('blocks/Census2010Block.shp')
BGs = BGs.to_crs(epsg = 4326)

city = ox.gdf_from_place('Stockton, California, USA')

BGs = gpd.overlay(BGs, city, how='intersection')

BGs.shape
```

```python
walkingIsochrones = pd.read_csv('isochrone_polys_walk.csv')
walkingIsochrones.set_index("Unnamed: 0",inplace=True)
walkingIsochrones.set_index(BGs.GEOID10,inplace=True)
```

```python
travel_speed_walk = 4.5 #walking speed in km/hour
G_walk = ox.graph_from_place('Stockton, California, USA', network_type='walk')
```

```python
geometry = [Point(xy) for xy in zip(stops.stop_lon, stops.stop_lat)]
stops = stops.drop(['stop_lon', 'stop_lat'], axis=1)
crs = {'init': 'epsg:4326'}
stops_gdf = GeoDataFrame(stops, crs=crs, geometry=geometry)
```

```python
trip_times = [1, 3, 5, 10, 15, 20, 25, 30]
```

```python
# make the isochrone polygons - WALK
isochrone_polys_walk_stops = pd.DataFrame(columns=[30,25,20,15,10,5,3,1],index=stops_gdf.stop_id)
for i in range(0, len(stops_gdf)):
    print(str(i+1) + " out of " + str(len(stops_gdf)))
    isochrone_polys = []
    center_node = ox.get_nearest_node(G_walk, (stops_gdf.geometry[i].y, stops_gdf.geometry[i].x))
    for trip_time in sorted(trip_times, reverse=True):
        subgraph = nx.ego_graph(G_walk, center_node, radius=trip_time, distance='time')
        node_points = [Point((data['x'], data['y'])) for node, data in subgraph.nodes(data=True)]
        bounding_poly = gpd.GeoSeries(node_points).unary_union.convex_hull
        isochrone_polys_walk_stops.loc[stops_gdf.stop_id[i],trip_time] = bounding_poly
```

```python
isochrone_polys_walk_stops.to_csv('transit_stop_walkingIsochrones.csv')
```

```python
def union(time, isos, iso_30, iso_25, iso_20, iso_15, iso_10, iso_5, iso_3, iso_1):
    if time == 0:
        iso_30 = iso_30.union(isos[30])
    elif time <= 5 and time > 0:
        iso_30 = iso_30.union(isos[25])
    elif time <= 10 and time > 5:
        iso_30 = iso_30.union(isos[20])
    elif time <= 15 and time > 10:
        iso_30 = iso_30.union(isos[15])
    elif time <= 20 and time > 15:
        iso_30 = iso_30.union(isos[10])
    elif time <= 25 and time > 20:
        iso_30 = iso_30.union(isos[5])
    elif time <= 27 and time > 22:
        iso_30 = iso_30.union(isos[3])
    elif time <= 29 and time > 24:
        iso_30 = iso_30.union(isos[1])
    
    if time == 0:
        iso_25 = iso_25.union(isos[25])
    elif time <= 5 and time > 0:
        iso_25 = iso_25.union(isos[20])
    elif time <= 10 and time > 5:
        iso_25 = iso_25.union(isos[15])
    elif time <= 15 and time > 10:
        iso_25 = iso_25.union(isos[10])
    elif time <= 20 and time > 15:
        iso_25 = iso_25.union(isos[5])
    elif time <= 22 and time > 17:
        iso_25 = iso_25.union(isos[3])
    elif time <= 24 and time > 19:
        iso_25 = iso_25.union(isos[1])
    
    
    if time == 0:
        iso_20 = iso_20.union(isos[20])
    elif time <= 5 and time > 0:
        iso_20 = iso_20.union(isos[15])
    elif time <= 10 and time > 5:
        iso_20 = iso_20.union(isos[10])
    elif time <= 15 and time > 10:
        iso_20 = iso_20.union(isos[5])
    elif time <= 17 and time > 12:
        iso_20 = iso_20.union(isos[3])
    elif time <= 19 and time > 14:
        iso_20 = iso_20.union(isos[1])

    if time == 0:
        iso_15 = iso_15.union(isos[15])
    elif time <= 5 and time > 0:
        iso_15 = iso_15.union(isos[10])
    elif time <= 10 and time > 5:
        iso_15 = iso_15.union(isos[5])
    elif time <= 12 and time > 7:
        iso_15 = iso_15.union(isos[3])
    elif time <= 14 and time > 9:
        iso_15 = iso_15.union(isos[1])

    if time == 0:
        iso_10 = iso_10.union(isos[10])
    elif time <= 5 and time > 0:
        iso_10 = iso_10.union(isos[5])
    elif time <= 7 and time > 2:
        iso_10 = iso_10.union(isos[3])
    elif time <= 9 and time > 4:
        iso_10 = iso_10.union(isos[1])

    if time == 0:
        iso_5 = iso_5.union(isos[5])
    elif time <= 2 and time > 1:
        iso_5 = iso_5.union(isos[3])
    elif time <= 4 and time > 2:
        iso_5 = iso_5.union(isos[1])

    if time == 0:
        iso_3 = iso_3.union(isos[3])
    elif time <= 3 and time < 0:
        iso_3 = iso_3.union(isos[1])

    if time <= 1:
        iso_1 = iso_1.union(isos[1])
```
